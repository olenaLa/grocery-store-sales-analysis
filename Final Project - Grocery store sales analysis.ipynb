1. Який загальний дохід магазину за кожен рік?

```python
yearly_revenue = df_grocery.groupby('Year')['Gross Bill Amount'].sum()
pd.options.display.float_format = '{:.0f}'.format
yearly_revenue_df = yearly_revenue.reset_index()
yearly_revenue_df.columns = ['Рік', 'Загальний валовий дохід']

print(yearly_revenue_df)
```

2. Як змінюється дохід магазину в різні місяці?

```python
monthly_revenue = df_grocery.groupby('Month')['Gross Bill Amount'].sum()

plt.figure(figsize=(10, 6))
monthly_revenue.plot(kind='line', color='green', marker='o', linestyle='-')
plt.title("Store's revenue variation across months")
plt.xlabel("Month")
plt.ylabel("Total Revenue")
plt.xticks(range(1, 13))
plt.grid(True)
plt.show()
```

**Висновок**: Виходячи зі спостереження, що грудень (12-й місяць) і січень (1-й місяць) мають вищі доходи, магазину слід розглянути можливість запуску спеціальних акцій і маркетингових кампаній протягом цих місяців, щоб отримати вигоду від збільшення витрат клієнтів під час святкового сезону та початку року, потенційно підвищуючи загальну прибутковість. Крім того, магазин повинен забезпечити наявність достатнього запасу товарів та персоналу для задоволення підвищеного попиту в цей період.

3. Які найпопулярніші способи оплати обирають клієнти?

```python
unique_payment_modes = df_grocery['Payment Mode'].unique()

print("Unique Payment Modes:", unique_payment_modes)
```

```python
payment_mode_counts = df_grocery['Payment Mode'].value_counts()

custom_colors = {
    'Card': 'goldenrod',
    'Cash': 'limegreen',
    'Wallet': 'turquoise'
}

plt.figure(figsize=(10, 6))
for payment_mode, color in custom_colors.items():
    subset = payment_mode_counts[payment_mode_counts.index == payment_mode]
    plt.bar(subset.index, subset.values, color=color, label=payment_mode)

plt.title("Most popular payment modes")
plt.xlabel("Payment mode")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**Висновок**: На основі даних і бачення, що оплата готівкою ще поширена магазин може проводити рекламні акції, які заохочуватимуть клієнтів, які платять готівкою, спробувати цифрові методи оплати.

4. Яка категорія товарів приносить найбільший дохід магазину?

```python
category_revenue = df_grocery.groupby('Category')['Gross Bill Amount'].sum()

highest_revenue_category = category_revenue.idxmax()
total_highest_revenue = category_revenue.max()

print(f"Категорія з найбільшим доходом це'{highest_revenue_category}' з загальним доходом {total_highest_revenue:.2f}$.")
```

5. Чи є різниця в купівельній поведінці покупців чоловічої та жіночої статі?

```python
male_customers = df_grocery[df_grocery['Gender'] == 'Male']
female_customers = df_grocery[df_grocery['Gender'] == 'Female']
selected_columns = ['Gross Bill Amount']
gender_describe = df_grocery.groupby('Gender')[selected_columns].describe()
print(gender_describe)

plt.figure(figsize=(10, 6))
plt.hist([male_customers['Gross Bill Amount'], female_customers['Gross Bill Amount']], bins=20, color=['blue', 'pink'], alpha=0.7, label=['Male', 'Female'])
plt.title("Purchase behavior by gender")
plt.xlabel("Gross Bill Amount")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```

**Висновок**: необхідно розробляти цільові маркетингові кампанії та рекламні акції, які спеціально обслуговують уподобання жінок-клієнтів і купівельні звички. 

6. Який середній відсоток прибутку магазину?

```python
average_profit_margin_percentage = (df_grocery['Profit Margin'] / df_grocery['Gross Bill Amount']).mean() * 100
print(f"Середній відсоток прибутку магазину складає: {average_profit_margin_percentage:.2f}%")
```

7. Дослідимо залежність між суми валого прибутку і операційними витратами:

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df_grocery, x='Operating Cost', y='Gross Bill Amount', alpha=0.5)
sns.regplot(x='Operating Cost', y='Gross Bill Amount', data=df_grocery, line_kws={'color':'red', 'lw':'1'}, scatter_kws={'alpha': 0.1});
plt.title("Operating Cost vs. Gross Bill Amount")
plt.xlabel("Operating Cost")
plt.ylabel("Gross Bill Amount")
plt.grid(True)
plt.show()
```

**Висновок**: враховуючи концентрацію залежності у діапазоні 10$ на осі операційних витрат і 250$ на осі суми валового рахунку, можливо, доцільно зосередитися на оптимізації та впорядкуванні операцій у цьому діапазоні. Магазину слід шукати способи скорочення операційних витрат, зберігаючи або збільшуючи суми валових рахунків у цій «ефективній» зоні. Також слід провести аналіз викидів, які можуть представляти унікальні проблеми, які потребують особливої уваги.

8. Проведемо додатковий аналіз прибутковості шляхом окремого розрахунку маржі прибутку для операцій у зоні концентрації з попереднього аналізу та викидів

```python
concentration_zone = (df_grocery['Operating Cost'] <= 10) & (df_grocery['Gross Bill Amount'] <= 250)

concentration_data = df_grocery[concentration_zone]
outliers_data = df_grocery[~concentration_zone]

average_profit_margin_concentration = concentration_data['Profit Margin'].mean()
average_profit_margin_outliers = outliers_data['Profit Margin'].mean()

print(f"Середня рентабельність у зоні концентрації: {average_profit_margin_concentration:.2%}")
print(f"Середня рентабельність для викидів: {average_profit_margin_outliers:.2%}")
```

**Висновок**: Слід продовжувати зосереджуватися на оптимізації операцій і прибутковості в зоні концентрації, оскільки вона, становить значну частину бізнесу магазину.

9. Як змінюються операційні витрати магазину у відсотках від доходу з роками?

```python
yearly_data = df_grocery.groupby('Year').agg({'Operating Cost': 'sum', 'Product Cost': 'sum', 'Gross Bill Amount': 'sum'})
yearly_data['Operating Cost Percentage'] = (yearly_data['Operating Cost'] / yearly_data['Gross Bill Amount']) * 100
yearly_data['Net Profit'] = yearly_data['Gross Bill Amount'] - (yearly_data['Operating Cost'] + yearly_data['Product Cost'])

print(yearly_data[['Net Profit', 'Operating Cost Percentage']])
```

**Висновок**: поєднання зростання чистого прибутку та стабільного відсотка операційних витрат свідчить про те, що магазин ефективно керував своїми витратами, збільшуючи дохід. Це позитивний знак для фінансового стану магазину.

11. Подивимось на динаміку суми чистого доходу по годинам в країнах. 

```python
df_grocery['Hour'] = df_grocery['Time'].dt.hour
net_bill_amount_country_pivot = pd.pivot_table(
    df_grocery,
    values='Net Bill Amount',
    index='Hour',
    columns='Country',
    aggfunc='sum'
)
print(net_bill_amount_country_pivot)
```

```python
color_palette = sns.color_palette('deep', len(net_bill_amount_country_pivot.columns))
plt.figure(figsize=(12, 6))
for i, country in enumerate(net_bill_amount_country_pivot.columns):
    plt.plot(net_bill_amount_country_pivot.index, net_bill_amount_country_pivot[country], marker='s', color=color_palette[i], label=country)

plt.title('Hourly Net Bill Amount by Country')
plt.xlabel('Hour')
plt.ylabel('Net Bill Amount')
plt.grid(True)
plt.legend()
plt.xticks(range(24))
plt.show()
```

**Висновок**: Модемо помітити коли продажі найвищі(в Філіпінах і Малазії ця тенденція явно припадає на кінец дня), і можемо виділити більше ресурсів або рекламних акцій у цей час, щоб максимізувати дохід.
Розуміння відмінностей по країнам може сформувати маркетингові та операційні стратегії, адаптовані до кожної країни.
